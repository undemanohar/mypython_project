```python
import openai

# Set up your OpenAI API key
openai.api_key = ''

def get_response(prompt):
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=150,
        n=1,
        stop=None,
        temperature=0.7,
    )
    return response.choices[0].text.strip()

def chat():
    print("Hello! I am a chatbot. Type 'exit' to end the conversation.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            print("Goodbye!")
            break
        response = get_response(user_input)
        print("Bot:", response)

if __name__ == "__main__":
    chat()
```

```python
import openai

# Set up your OpenAI API key
openai.api_key = ''

def get_response(prompt):
    response = openai.ChatCompletion.create(  # Use ChatCompletion instead of Completion
        model="gpt-3.5-turbo",  # Specify a chat model
        messages=[{"role": "user", "content": prompt}],  # Use a messages list
        max_tokens=150,
        n=1,
        stop=None,
        temperature=0.7,
    )
    return response.choices[0].message.content.strip()  # Extract content from message

def chat():
    print("Hello! I am a chatbot. Type 'exit' to end the conversation.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            print("Goodbye!")
            break
        response = get_response(user_input)
        print("Bot:", response)

if __name__ == "__main__":
    chat()
```
